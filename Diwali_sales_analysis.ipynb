```python
ax = sns.countplot( x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x ='Gender', y='Amount', data = sales_gen)
```

From the above graph it is clearly visible that the number of female buyers is more and the purchasing power of females are also greater than men

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace = True)
```

```python
ax = sns.countplot( data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y= 'Amount', data= sales_age) 
```

From the above graph, it is clearly visible that most of the customers are from the age group 26-35yrs Female

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State', y= 'Orders')
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot( data = sales_state, x = 'State', y = 'Amount')
```

Frome the graph we can clearly see that most of the sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively


```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(7, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_states, x = 'Marital_Status', y = 'Amount', hue= 'Gender')
```

From the above graph its clearly visible that the purchasing power of women's are more as they have done the most shopping. 

```python
ax = sns.countplot(data = df, x = 'Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot( data = sales_states, x = 'Occupation', y = 'Amount')
```

From the above graphs it is clearly visible that the most buyers are working in IT,Aviation and Healthcare sector

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Product_Category')
sns.set(rc={'figure.figsize':(25,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot( data = sales_states, x = 'Product_Category', y = 'Amount')
```

From above graphs it is visible that most of the sold products are from Food, Clothing Footwear and Electronics category

```python
sales_states = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot( data = sales_states, x = 'Product_ID', y = 'Orders')
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
